# Pythonでやる三角関数サンプル

```python
# ライブラリのインポート
import numpy as np

# theta = Θ = シータ 
theta = float(input("実数値を入力：")) # deg値で角度を入力させる

r = np.radians(theta) # rad値に角度を変換する
print(f"{theta} のラジアン(rad)値：{r:.3f}") # 有効桁数3桁でrad値を表示

print(f"sin{theta} = {np.sin(r):.3f}") # 有効桁数3桁でsin値を表示
print(f"cos{theta} = {np.cos(r):.3f}") # 有効桁数3桁でcos値を表示
print(f"tan{theta} = {np.tan(r):.3f}") # 有効桁数3桁でtan値を表示
```

```python
# グラフ出力サンプル
import numpy as np
import matplotlib.pyplot as plt

# 波形の周期
n = 3

deg = np.array([i for i in range(360*n + 1)])
rad = np.radians(deg)

t_rad = [r for r in rad]
for i in np.where(deg / 90 % 2 == 1)[0]:
    t_rad[i] = np.nan

sin = np.sin(rad)
cos = np.cos(rad)
tan = np.tan(t_rad)



plt.plot(deg,sin,label="sin")
plt.plot(deg,cos,label="cos")
# plt.plot(deg,tan,label="tan")
plt.legend()

plt.show()
```
