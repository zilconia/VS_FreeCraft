# Pythonでやる三角関数サンプル

```python
# ライブラリのインポート
import numpy as np

# theta = Θ = シータ 
theta = float(input("実数値を入力：")) # deg値で角度を入力させる

r = np.radians(theta) # rad値に角度を変換する
print(f"{theta} のラジアン(rad)値：{r:.3f}") # 有効桁数3桁でrad値を表示

print(f"sin{theta} = {np.sin(r):.3f}") # 有効桁数3桁でsin値を表示
print(f"cos{theta} = {np.cos(r):.3f}") # 有効桁数3桁でcos値を表示
print(f"tan{theta} = {np.tan(r):.3f}") # 有効桁数3桁でtan値を表示
```

```python
# グラフ出力サンプル
import numpy as np
import matplotlib.pyplot as plt

# 波形の周期
n = 2

deg = np.array([i for i in range(360*n + 1)])
rad = np.radians(deg)

t_rad = [r for r in rad] # tan波の計算用にrad値をコピー
for i in np.where(deg / 90 % 2 == 1)[0]: 
    t_rad[i] = np.nan 

sin = np.sin(rad) # sin波の計算
cos = np.cos(rad) # cos波の計算
tan = np.tan(t_rad) # tan波の計算

plt.plot(deg,sin,marker="o",markevery=30,label="sin") # sin波の表示
plt.plot(deg,cos,marker="o",markevery=30,label="cos") # cos波の表示
# plt.plot(deg,tan,label="tan") # tan波の表示

plt.title("sin,cos,tan wave") # グラフタイトル

plt.xticks([i for i in range(0,360*n + 1,30*n)]) # x軸の目盛りを30*n度ごとに設定
plt.legend() # 凡例の表示

plt.show() # グラフの表示
```
